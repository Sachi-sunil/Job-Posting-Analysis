```python
import seaborn as sns
import matplotlib.pyplot as plt

top_jobs = df['Job_Ttl'].value_counts().head(10)
sns.barplot(y=top_jobs.index, x=top_jobs.values)
plt.title("Top 10 Job Titles")
plt.xlabel("Number of Postings")
plt.ylabel("Job Title")
plt.show()


sns.countplot(x='is_remote', data=df)
plt.title("Remote vs On-site Jobs")
plt.xticks([0, 1], ['On-site', 'Remote'])
plt.show()
```

```python
df.drop(columns=['Co_Pg_Lstd', 'py_lstd', 'app_is_off'], inplace=True)
```

```python
top_titles = df['Job_Ttl'].value_counts().head(10)
sns.barplot(y=top_titles.index, x=top_titles.values, palette="coolwarm")
plt.title("Top 10 Job Titles")
plt.xlabel("Number of Postings")
plt.ylabel("Job Titles")
plt.show()
```

```python
top_companies = df['Co_Nm'].value_counts().head(10)
sns.barplot(y=top_companies.index, x=top_companies.values, palette="magma")
plt.title("Top Hiring Companies")
plt.xlabel("Number of Postings")
plt.ylabel("Company")
plt.show()
```

```python
remote_jobs = df['is_remote'].value_counts()
sns.barplot(x=remote_jobs.index, y=remote_jobs.values)
plt.title("Remote vs On-site Jobs")
plt.xticks([0, 1], ['On-site', 'Remote'])
plt.ylabel("Count")
plt.show()
```

```python
top_salary_jobs = df.groupby('Job_Ttl')['med_sal'].mean().sort_values(ascending=False).head(10)
top_salary_jobs.plot(kind='barh', color='green')
plt.title("Top 10 Job Titles by Median Salary")
plt.xlabel("Median Salary")
plt.ylabel("Job Title")
plt.show()
```

```python
# Work Type
df['wrk_typ'].value_counts().plot(kind='bar', color='skyblue', title='Work Type Distribution')
plt.ylabel("Number of Postings")
plt.show()

# Experience Level
df['xp_lvl'].value_counts().plot(kind='bar', color='orange', title='Experience Level Distribution')
plt.ylabel("Number of Postings")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['med_sal'], bins=30, kde=True, color='teal')
plt.title('Distribution of Median Salary')
plt.xlabel('Median Salary')
plt.ylabel('Number of Job Postings')
plt.show()
```

```python
remote_salary = df.groupby('is_remote')['med_sal'].mean()
remote_salary.index = ['On-site', 'Remote']

remote_salary.plot(kind='bar', color=['#FF9999','#66B3FF'])
plt.title('Average Median Salary: Remote vs On-site')
plt.ylabel('Average Median Salary')
plt.xticks(rotation=0)
plt.show()
```

```python
xp_salary = df.groupby('xp_lvl')['med_sal'].mean().sort_values()
xp_salary.plot(kind='barh', color='slateblue')
plt.title('Median Salary by Experience Level')
plt.xlabel('Median Salary')
plt.ylabel('Experience Level')
plt.show()
```

```python
top_locations = df['loc'].value_counts().head(10)
sns.barplot(y=top_locations.index, x=top_locations.values, palette='viridis')
plt.title("Top 10 Job Locations")
plt.xlabel("Number of Postings")
plt.ylabel("Location")
plt.show()
```

```python
sns.scatterplot(data=df, x='views', y='base_comp', alpha=0.5)
plt.title('Views vs Compensation')
plt.xlabel('Views')
plt.ylabel('Base Compensation')
plt.show()
```
